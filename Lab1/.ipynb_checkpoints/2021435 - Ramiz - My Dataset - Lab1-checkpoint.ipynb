# Dealing with missing values

```python
missing_cols = []
for col in df.columns:
    if len(df[col].isnull().value_counts().values)==2:
        missing_cols.append(col)
```

```python
print(missing_cols)
```

```python
for col in missing_cols:
    display(df[col])
```

```python
for col in missing_cols:
    print(col)
    print(df[df[col].isnull()==True].shape[0])
    display(df[df[col].isnull()==True])
```

```python
pd.set_option('display.max_columns',None)
```

```python
missing_cols_48 = []
rest_cols = []
count=0
for col in missing_cols:
#     print(col,end=" ")
#     print(df[df[col].isnull()==True].shape[0])
    if (df[df[col].isnull()==True].shape[0]==48):
        count+=1
        missing_cols_48.append(col)
    else:
        rest_cols.append(col)
print(count)
```

```python
len(missing_cols)
```

```python
missing_cols_48
```

```python
rest_cols
```

```python
for col in rest_cols:
    print(col)
    print(df[df[col].isnull()==True].shape[0])
    display(df[df[col].isnull()==True])
```

```python
# Club - Club Not Mentioned
# Position - Position Not Mentioned
# Jersey Number - 0
# Joined - Jan 1, 1970
# Loaned From - Not Loaned
# Contract Valid Until - str(2100)
# Release Clause - str(0)
```

```python
replace_missing_vals = {'Club':'Club Not Mentioned','Position':'Position Not Mentioned',
                       'Jersey Number':0,'Joined':'Jan 1, 1970',
                        'Loaned From':'Not Loaned','Contract Valid Until':'2100',
                       'Release Clause':'0'}
```

```python
for col in replace_missing_vals:
    print(col,end=" ")
    print(replace_missing_vals[col])
    df[col].fillna(replace_missing_vals[col],inplace=True)
```

```python
df.dropna(inplace=True)
```
